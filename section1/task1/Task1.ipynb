```python
import pandas as pd
import requests
from datetime import datetime
from datetime import timezone
```

```python
class CryptoCompare:
    
    def timeStamp(self,date):
        timestamp = date.replace(tzinfo=timezone.utc).timestamp()
        return int(timestamp)
    
    def getData(self,fsym,tsym,start_time,end_time,e):
        http = "https://min-api.cryptocompare.com/data/v2/histohour?fsym="+fsym+"&tsym="+tsym+"&limit=1&toTs="+end_time+"&e="+ e
        r = requests.get(http)
        print(r)
        data = r.json()
        value = data['Data']['Data']
        df = pd.DataFrame(value, columns=['close','high', 'low','open','volumefrom','volumeto','time'])
        end_time = str(data['Data']['TimeFrom'])
        while (data['Data']['TimeFrom'] > int(start_time)):
            http = "https://min-api.cryptocompare.com/data/v2/histohour?fsym="+fsym+"&tsym="+tsym+"&limit=2000&toTs="+end_time+"&e="+ e
            r = requests.get(http)
            data = r.json()
            value = data['Data']['Data']
            df1 = pd.DataFrame(value, columns=['close','high', 'low','open','volumefrom','volumeto','time'])
            df = df1.append(df)
            end_time = str(data['Data']['TimeFrom'])
        df = df.reset_index(drop=True)
        df = df[df['time']>=int(start_time)]
        df = df.reset_index(drop=True)
        df['time'] = pd.to_datetime(df['time'],unit='s')
        df = df.rename(columns={"volumefrom": "volume", "volumeto": "baseVolume","time":"datetime"})
        df = df.drop_duplicates()
        df.to_csv('data.csv',index=False)
        return df
```

```python
a = CryptoCompare()
print(a.timeStamp(datetime(2020,4,1)))
print(a.timeStamp(datetime(2017,4,1)))
a.getData('BTC','USDT','1491004800','1585699200','binance')
```
